# Plotting functions
based on [official tutorial](http://seaborn.pydata.org/tutorial/regression.html)
## Visualizing linear relationships

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
plt.rc('font', family='DejaVu Sans') 

np.random.seed(sum(map(ord, 'regression')))

tips = sns.load_dataset('tips')
```

```python
sns.regplot(x='total_bill', y='tip', data=tips);
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips);
```

```python
sns.lmplot(x='size', y='tip', data=tips);
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_jitter=.1);
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.mean);
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.tail()
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query('dataset == "I"'),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query('dataset == "I"'));
```

### High Order

```python
sns.lmplot(x='x', y='y', data=anscombe.query('dataset == "II"'),
           ci=None, scatter_kws={'s': 80});
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

### Outlier

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"));
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"), robust=True, ci=None);
```

### Logistic regression

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           y_jitter=.03);
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03);
```

### Lowess smoother
https://en.wikipedia.org/wiki/Local_regression

```python
sns.lmplot(x="total_bill", y="tip", data=tips,
           lowess=True);
```

### Regression Residuals

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'III'"));
```

### Combine few variables

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"], palette="Set1");
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips);
```

```python
sns.lmplot(x="x", y="y", col="dataset", col_wrap=2, data=anscombe);
```

```python
f, ax = plt.subplots(figsize=(5, 6))
sns.regplot(x="total_bill", y="tip", data=tips, ax=ax);
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips,
           col_wrap=2, size=3);
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips,
           aspect=.5);
```

### With other plots

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],
             size=5, aspect=.8, kind="reg");
```

```python
sns.pairplot(tips, vars=["total_bill", 'tip', "size"], kind='reg');
```

```python
sns.pairplot(tips, vars=["total_bill", 'tip', "size"], hue='smoker', kind='reg');
```
