## Controlling figure aesthetics

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, 'aesthetics')))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
import seaborn as sns
# list available fonts: [f.name for f in matplotlib.font_manager.fontManager.ttflist]
mpl.rc('font', family='DejaVu Sans') 
sinplot()
```

### Different stylies

```python
sns.set_style('whitegrid')
mpl.rc('font', family='DejaVu Sans') 
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
mpl.rc('font', family='DejaVu Sans') 
sinplot()
```

```python
sns.set_style('white')
mpl.rc('font', family='DejaVu Sans') 
sinplot()
```

```python
sns.set_style('ticks')
mpl.rc('font', family='DejaVu Sans') 
sinplot()
```

```python
sinplot()
sns.despine()
```

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
```

```python
sns.set_style("whitegrid")
mpl.rc('font', family='DejaVu Sans') 
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    mpl.rc('font', family='DejaVu Sans') 
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.set_context('paper')
plt.figure(figsize=(8, 8))
plt.subplot(4,1,1)
sinplot()
sns.set_context('talk')
plt.figure(figsize=(8, 8))
plt.subplot(4,1,2)
sinplot()
sns.set_context('poster')
plt.figure(figsize=(8, 8))
plt.subplot(4,1,3)
sinplot()
sns.set_context('notebook')
plt.figure(figsize=(8, 8))
plt.subplot(4,1,4)
sinplot()
```

```python
sns.palplot(sns.color_palette('Blues_d', 8))
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=300).T
cmap = sns.cubehelix_palette(light=1, start=.5, rot=-.75, as_cmap=True)
ax = sns.kdeplot(x, y, cmap=cmap, shade=True)
```

```python
sns.axes_style("darkgrid")
mpl.rc('font', family='DejaVu Sans') 
ax = plt.plot(x, y, '.')
```

```python
iris = sns.load_dataset("iris")
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,
                 cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
                cmap="Blues", shade=True, shade_lowest=False)
```
